```python
import matplotlib.pyplot as plt

x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

dx = x2 - x1
dy = y2 - y1

x = x1
y = y1

X = []
Y = []


p = 2*dy - dx

while x <= x2:
    X.append(x)
    Y.append(y)

    if p >= 0:
        y = y + 1
        p = p + 2*dy - 2*dx
    else:
        p = p + 2*dy

    x = x + 1

plt.plot(X, Y, marker='o')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bresenham Line Drawing Algorithm")
plt.grid(True)
plt.show()
```
